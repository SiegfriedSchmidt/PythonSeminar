## Задача Matplotlib 3 (10 баллов)

**Требуется**

1. самостоятельно придумать непрерывную функцию $f(x, y)$, нелинейно зависящую от $x$ и $y$.
2. нарисовать контурную карту функции $f(x, y)$ на выбранной самостоятельно прямоугольной области плоскости $x-y$.
3. добавить подписи к осям
4. добавить заголовок к рисунку
5. отобразить цветовую шкалу
6. cохранить изображение в формате `pdf` и разрешении `300 dpi`

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return np.sin(x) * np.cos(y) + 0.05 * (x * x + y * y)


x = np.linspace(-6.0, 6.0, 100)
y = np.linspace(-6.0, 6.0, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig, axes = plt.subplots()

cont = axes.contour(X, Y, Z, levels=30, cmap='Spectral')
axes.set_title(f'f(x, y) = sin(x) * cos(x) + 0.05 * (x ^ 2 + y ^ 2)')
axes.set_xlabel("X")
axes.set_ylabel("Y")

plt.colorbar(cont, label="Z")
plt.savefig('./contour_plot.pdf', dpi=300)
````

# Задача Numpy* (15 баллов)

**Дано**

В файлах `nodes.npy` и `triangles.npy` содержатся массивы `Numpy`, которые можно загрузить при помощи функции `np.load`:
- `nodes` - массив (N, 2), содержащий координаты точек (float)
- `tris` - массив (M, 3), содержащий тройки индексов (i, j, k) такие, что точки с индексами i, j, k из массива `nodes` образуют треугольник.

**Требуется**

1. Вычислить минимальное и максимальное значение среди всех треугольников:
    1. длин сторон
    2. периметров
    3. площадей
    4. углов

**Оценивание**

- [5 баллов] допускаются любые циклы и их заменители
- [10 баллов] не допускаются никакие циклы и их заменители; использовать только возможности `Numpy`

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

nodes = np.load("./nodes.npy")
tris_indexes = np.load("./triangles.npy").astype(np.int64)

tris = nodes[tris_indexes]
lengths = np.dstack((
    np.linalg.norm(tris[:, 0] - tris[:, 1], axis=1),
    np.linalg.norm(tris[:, 1] - tris[:, 2], axis=1),
    np.linalg.norm(tris[:, 2] - tris[:, 0], axis=1)
)).squeeze(axis=0)

perimeters = np.sum(lengths, axis=1)
semi_perimeters = perimeters / 2

areas = np.sqrt(semi_perimeters * (semi_perimeters - lengths[:, 0]) * (semi_perimeters - lengths[:, 1]) * (semi_perimeters - lengths[:, 2]))

angles = np.rad2deg(np.dstack((
    np.arccos((lengths[:, 0] ** 2 + lengths[:, 1] ** 2 - lengths[:, 2] ** 2) / (2 * lengths[:, 0] * lengths[:, 1])),
    np.arccos((lengths[:, 1] ** 2 + lengths[:, 2] ** 2 - lengths[:, 0] ** 2) / (2 * lengths[:, 1] * lengths[:, 2])),
    np.arccos((lengths[:, 2] ** 2 + lengths[:, 0] ** 2 - lengths[:, 1] ** 2) / (2 * lengths[:, 2] * lengths[:, 0])),
)).squeeze(axis=0))

print(f"Минимальная длина: {lengths.min()}")
print(f"Максимальная длина: {lengths.max()}")

print(f"Минимальный периметр: {perimeters.min()}")
print(f"Максимальный периметр: {perimeters.max()}")

print(f"Минимальная площадь: {areas.min()}")
print(f"Максимальная площадь: {areas.max()}")

print(f"Минимальный угол: {angles.min()}")
print(f"Максимальный угол: {angles.max()}")

````

# Задача Pandas* (15 баллов)

В файле `fin_v2.csv` находятся записи о доходах и расходах физического лица.

Требуется рассчитать:

1. среднемесячный доход для каждого года
2. среднемесячный расход для каждого года
3. состояние счета физического лица на каждый день и изобразить график (считать, что в начальный момент времени состояние счета было нулевым)
4. даты, когда счет становился отрицательным
5. 5 дат, когда произошли самые большие траты и их категории и величины
6. построить прогнозы (линейный и квадратичный) состояния счета через 2 года после последней записи
7. выбросить из данных две самых больших траты и снова построить прогнозы

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from datetime import datetime, timedelta

df = pd.read_csv('./fin_v2.csv', sep=',', parse_dates=["ВРЕМЯ"])
print(df['КАТЕГОРИЯ'].unique())

incomes = df[df['КАТЕГОРИЯ'] == 'Доходы'].copy()
incomes['Год'] = incomes['ВРЕМЯ'].dt.year
incomes['Месяц'] = incomes['ВРЕМЯ'].dt.month
month_income = incomes.groupby(['Год', 'Месяц'])['СУММА'].sum()
year_income = month_income.groupby('Год').mean()

# print(year_income)
# year_income = df.groupby([month_income['ВРЕМЯ'].dt.to_period('Y')]).sum()
# print(year_income)
# print(df[df['КАТЕГОРИЯ'] == 'Доходы']['СУММА'])

print("Среднемесячный доход для каждого года:")
print(year_income)

expenses = df[df['КАТЕГОРИЯ'] != 'Доходы'].copy()
expenses['Год'] = expenses['ВРЕМЯ'].dt.year
expenses['Месяц'] = expenses['ВРЕМЯ'].dt.month
month_expense = expenses.groupby(['Год', 'Месяц'])['СУММА'].sum().abs()
year_expense = month_expense.groupby('Год').mean()
print("\nСреднемесячный расход для каждого года:")
print(year_expense)

all_dates = pd.date_range(start=df['ВРЕМЯ'].min().date(), end=df['ВРЕМЯ'].max().date(), freq='D')
day_transactions = df.groupby(df['ВРЕМЯ'].dt.date)['СУММА'].sum()
day_balance = day_transactions.reindex(all_dates, fill_value=0).cumsum()

day_balance.plot()
plt.title('Состояние счета по дням')
plt.xlabel('Дата')
plt.ylabel('Состояние счета')
plt.grid()
plt.show()

negative_dates = day_balance[day_balance < 0].index
print("\nДаты с отрицательным счетом:")
print(negative_dates)

most_expenses = expenses.nsmallest(5, 'СУММА')[['ВРЕМЯ', 'СУММА', 'КАТЕГОРИЯ']]
print("\nСамые большие траты:")
print(most_expenses)

min_date = df['ВРЕМЯ'].min()
df['Дни'] = (df['ВРЕМЯ'] - min_date).dt.days

x = df[['Дни']]
y = day_balance.reindex(df['ВРЕМЯ'].dt.date).values
model = LinearRegression()
model.fit(x, y)

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
model_poly = LinearRegression()
model_poly.fit(x_poly, y)

last_date = df['ВРЕМЯ'].max()
future_date = last_date + timedelta(days=365*2)
days_future = (future_date - min_date).days

linear_pred = model.predict([[days_future]])
poly_pred = model_poly.predict(poly.transform([[days_future]]))

print(f"\nПрогноз счета на {future_date.date()}:")
print(f"Линейный прогноз: {linear_pred[0]:.2f}")
print(f"Квадратичный прогноз: {poly_pred[0]:.2f}")

df_2 = df.copy()
biggest_expense_indices = expenses.nsmallest(2, 'СУММА').index
df_2 = df_2.drop(biggest_expense_indices)

daily_transactions_2 = df_2.groupby(df_2['ВРЕМЯ'].dt.date)['СУММА'].sum()
daily_balance_2 = daily_transactions_2.reindex(all_dates, fill_value=0).cumsum()

x_2 = df_2[['Дни']]
y_2 = daily_balance_2.reindex(df_2['ВРЕМЯ'].dt.date).values
model_2 = LinearRegression()
model_2.fit(x_2, y_2)

x_poly_2 = poly.fit_transform(x_2)
model_poly_2 = LinearRegression()
model_poly_2.fit(x_poly_2, y_2)

linear_pred_2 = model_2.predict([[days_future]])
poly_pred_2 = model_poly_2.predict(poly.transform([[days_future]]))

print(f"\nПрогноз счета на {future_date.date()} (без 2 самых больших трат):")
print(f"Линейный прогноз: {linear_pred_2[0]:.2f}")
print(f"Квадратичный прогноз: {poly_pred_2[0]:.2f}")

# график
plot_df = pd.DataFrame({
    'Дата': day_balance.index,
    'Реальный баланс': day_balance.values
})

future_dates = pd.date_range(start=min_date, end=future_date, freq='D')
future_days = np.array([(x - min_date).days for x in future_dates]).reshape(-1, 1)

linear_projection = model.predict(future_days)
poly_projection = model_poly.predict(poly.transform(future_days))

forecast_df = pd.DataFrame({
    'Дата': future_dates,
    'Линейный прогноз': linear_projection,
    'Квадратичный прогноз': poly_projection
})

combined_df = pd.concat([plot_df, forecast_df])
plt.figure(figsize=(14, 7))
plt.plot(combined_df['Дата'], combined_df['Реальный баланс'],label='Реальный баланс', color='blue', linewidth=2)
plt.plot(combined_df['Дата'], combined_df['Линейный прогноз'],label='Линейный прогноз', color='green', linestyle='--')
plt.plot(combined_df['Дата'], combined_df['Квадратичный прогноз'],label='Квадратичный прогноз', color='red', linestyle='--')
plt.title('Состояние счета')
plt.xlabel('Дата')
plt.ylabel('Состояние счета')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()

plot_df_2 = pd.DataFrame({
    'Дата': daily_balance_2.index,
    'Реальный баланс': daily_balance_2.values
})

linear_projection_2 = model_2.predict(future_days)
poly_projection_2 = model_poly_2.predict(poly.transform(future_days))

forecast_df_2 = pd.DataFrame({
    'Дата': future_dates,
    'Линейный прогноз': linear_projection_2,
    'Квадратичный прогноз': poly_projection_2
})

combined_df_2 = pd.concat([plot_df_2, forecast_df_2])

plt.figure(figsize=(14, 7))
plt.plot(combined_df_2['Дата'], combined_df_2['Реальный баланс'],label='Реальный баланс', color='blue', linewidth=2)
plt.plot(combined_df_2['Дата'], combined_df_2['Линейный прогноз'],label='Линейный прогноз', color='green', linestyle='--')
plt.plot(combined_df_2['Дата'], combined_df_2['Квадратичный прогноз'],label='Квадратичный прогноз', color='red', linestyle='--')
plt.title('Состояние счета (без 2 самых больших трат)')
plt.xlabel('Дата')
plt.ylabel('Состояние счета')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()

````
